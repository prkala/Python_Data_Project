### Assignment Operators

```python
applicant_1 = 'Kelly'
applicant_2 = 'Luke'
```

```python
# we want to interchange applicants
# method 1

temp_var = applicant_1
applicant_1 = applicant_2
applicant_2 = temp_var

print(applicant_1, applicant_2)
```

```python
# method 2

applicant_1, applicant_2 = applicant_2, applicant_1

print(applicant_1, applicant_2)
```
