## **Sets** mainly designed for unique items

There is no indexing in sets

No specific order

```python
job_skills = {'tableau', 'sql', 'python', 'statistics'}
job_skills
```

```python
job_skills[1]  # no indexing in sets
```

```python
help(set)
```

```python
job_skills.add('looker')
```

```python
job_skills.add('sql')  # runs but doesn't add duplicate value
```

```python
# Main use of sets: removing duplicate values effectively from a list

skill_list = ['python', 'sql', 'statistics', 'tableau', 'python', 'sql', 'statistics', 'tableau']
```

```python
skill_list_no_dup = list(set(skill_list))
```

```python
skill_list_no_dup
```
