```python
!pip install datasets
import pandas as pd

from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
import matplotlib.pyplot as plt
```

```python
job_counts = df['job_title_short'].value_counts()
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Job Postings')
plt.ylabel('Count of Postings')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

```python
job_counts.plot(kind = 'bar')   # since it is a series, we don't need to specify x and y as the arguments
```

```python
job_counts.plot(kind = 'bar')
plt.title('Job Postings')
plt.ylabel('Count of Postings')
plt.xlabel('')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

```python
df.sort_values(by = 'job_posted_date', inplace = True)
df.plot(x = 'job_posted_date', y = 'salary_year_avg', kind = 'line')
```

## Median Salary vs Data Science Jobs

```python
# median salary by job_title
job_sal = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
job_sal.plot(kind = 'barh')
plt.title('Median Salary by Job Title')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.show()
```
