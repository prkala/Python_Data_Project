```python
role = "Data Analyst"
skill = "Python"
```

```python
# Write out: 'Role: Data Analyst'

'Role: ' + role
```

```python
help(str.format)
```

```python
# args: arguments
# kwargs: keyword arguments
```

```python
# Create the string: "Role: Data Analyst"

"Role: {}".format(role)
```

```python
# Create the string "Role: Data Analyst; Skill Required: Python"
# using format method

"Role: {}; Skill Required: {}".format(role, skill)
```

```python
# Using keyword arguments

"Role: {r}; Skill Required: {s}".format(r=role, s =skill)
```

### f-strings - formatted string literals

```python
role = 'Data Analyst'
skill = 'Python'
```

```python
f"Role: {role}; Skill Required: {skill}"
```

### %-formatting (printf-style string formatting)

```python
# not that important to learn; but should be aware of this

"Role: %s; Skill Required: %s" % (role, skill)
```
