Tuples are similar to lists but immutable

They are indexable like lists

```python
# more applicability in software engineering


lukes_skills = ('python', 'sql', 'statistics', 'tableau')
```

```python
lukes_skills
```

```python
lukes_skills[0]
```

```python
lukes_skills[:2]
```

```python
lukes_skills.append('r')
```

```python
help(tuple)
```

```python
lukes_skills = ('python', 'sql', 'statistics', 'tableau')
print(id(lukes_skills))
new_skills = ('excel', 'power bi')

skills = lukes_skills + new_skills
skills
```

```python
lukes_skills += new_skills
lukes_skills    # this creates a new tuple lukes_skills; it's not the old one
```

```python
# investigating the new tuple creation


id(lukes_skills)
```

```python
lukes_skills_list = ['python', 'sql', 'statistics', 'tableau']
new_skills = ['excel', 'power bi']

print(id(lukes_skills_list))

lukes_skills_list += new_skills

print(id(lukes_skills_list))
```
