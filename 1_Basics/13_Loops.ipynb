## For and While **Loops**

```python
def pyramid_print(num_levels):
  for i in range(num_levels+1):
    print(" "*(num_levels+1-i), end="")
    print("*"*(i*2-1))
```

```python
pyramid_print(6)
```

```python
for num in range(100):
  if num % 2 != 0:
    print(num, end=" ")
```

```python
numbers = [0,1,2,3,4,5]
```

```python
for number in numbers:
  print(number)
```

```python
characters = 'PYTHON'

for char in characters:
  print(char)
```

```python
import random

names = [
    "Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", "Hannah",
    "Ian", "Jasmine", "Kevin", "Laura", "Mike", "Nina", "Owen", "Paula",
    "Quincy", "Rachel", "Steve", "Tina", "Uma", "Victor", "Wendy", "Xander", "Yara"
]

years_exp = {name: random.randint(1, 15) for name in names}
```

```python
years_exp
```

```python
for person in years_exp:   # for printing keys
  print(person)
```

```python
help(dict)
```

```python
for value in years_exp.values():  # for printing values
  print(value)
```

```python
for value in years_exp.items():
  print(value)
```

```python
for key, value in years_exp.items():
  if value > 5:
    print(key, "has", value, "years of exp")
```

## While Loops

```python
i = 10
while i > 0:
  print("Data Nerd")
  i-=1
```

```python
while False:
  print("Hello")
```

```python
years_exp_new = {
    "Luke": 3,
    "Kelly": 6,
    "Ken": 4,
    "Alex": 3,
    "Mary": 7
}
```

```python
years = 0
index = 0

years_list = list(years_exp_new.items())   # creating a list out of the dictionary

while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, "has", years, "years of experience")
  index += 1
```
