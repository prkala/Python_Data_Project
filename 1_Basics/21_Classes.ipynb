```python
# Creating a class which mimics a list

class LukesList:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
```

```python
my_list = LukesList()
```

```python
# self is the first positional argument, base_salary is the second

class BaseSalary:
  def __init__(self, base_salary, bonus_rate = 0.1, symbol = '$'):
    self.base_sal = base_salary
    self.bonus_rt = bonus_rate
    self.symb = symbol
    self.total_sal = base_salary * (1 + bonus_rate)
    self.bonus = self.total_sal - base_salary

  def __repr__(self):
    return f"{self.symb}{self.base_sal:,.0f}"

  def show_salary(self):
    return f'{self.symb}{self.total_sal:,.0f}'

  def show_bonus(self):
    return f'{self.symb}{self.bonus:,.0f}'
```

```python
sal = BaseSalary(100000)
sal.base_sal, sal.bonus_rt, sal.symb
```
