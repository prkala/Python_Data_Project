```python
# Create a list of 1,000,000 salaries ranging from 50,000 to 150,000

import random

random.randint(50000, 150000)
```

```python
sal_list = [random.randint(50_000, 150_000) for x in range(1_000_000)]
```

```python
import statistics
```

```python
%%timeit

statistics.mean(sal_list)
```

```python
import numpy as np
```

```python
%%timeit

np.mean(sal_list)
```

```python
# Job titles
job_titles = np.array(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer'])

# Base salaries
base_salaries = np.array([60000, 80000, 75000, 90000])

# Bonus rates
bonus_rates = np.array([.05, .1, .08, .12])
```

```python
np.mean(total_salaries)
```

```python
# having a job title with salary and bonus
# we can use NaN instead of "0" or "None"

# Job titles
job_titles = np.array(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'AI Engineer'])

# Base salaries
base_salaries = np.array([60000, 80000, 75000, 90000, np.nan])

# Bonus rates
bonus_rates = np.array([.05, .1, .08, .12, np.nan])
```

```python
type(np.nan)
```

```python
total_salaries = base_salaries * (1 + bonus_rates)
total_salaries
```

```python
np.mean(total_salaries)   # mean can't handle nan values
```

```python
np.nanmean(total_salaries)
```
