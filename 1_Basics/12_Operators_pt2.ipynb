## Logical Operators

AND, OR and NOT

```python
job_work_from_home = True
job_health_insurance = True
job_no_degree_mention = False
```

```python
job_work_from_home and job_health_insurance and job_no_degree_mention
```

```python
job_work_from_home or job_no_degree_mention
```

```python
not True
```

```python
job_work_from_home and job_health_insurance and not job_no_degree_mention
```

## Identity Operators

IS and IS NOT

```python
sal_kelly = 100_000
sal_luke = 100_000
```

```python
sal_kelly == sal_luke
```

```python
sal_kelly is sal_luke    # is operator verifies whether the variables are using the same memory location for the value they are holding
```

```python
print(id(sal_kelly))
print(id(sal_luke))
```

```python
sal_kelly = sal_luke
sal_kelly is sal_luke
```
