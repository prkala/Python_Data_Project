Classes: a template for creating objects (records)

```python
help(int) # help on "class" integer
```

```python
job_title = 'Data Analyst'
job_location = 'United States'
job_salary = 90000
```

```python
class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
```

```python
JobPost(job_title, job_location, job_salary)   # creation of an instance
```

Attributes: Variables of an Object

**Attribute**: A field in a record, defined by its class.

```python
# title, location, salary are attributes of JobPost
```

```python
job_1 = JobPost(job_title, job_location, job_salary)
```

```python
# we can access the attributes of job_1 i.e. variables of the class
job_1.title
```

```python
job_1.location
```

Methods: The Functions of an Object

**Method**: A function defined inside a class that operates on its objects



```python

# example of a class and methods inside the class:

class JobPost:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
  def display_info(self):
    return print(f"JOB:      {self.title}\nLOCATION: {self.location}\nSALARY:   ${self.salary}")
```

```python
job_2 = JobPost(job_title, job_location, job_salary)
```

```python
job_2.title
```

```python
job_2.display_info()
```
