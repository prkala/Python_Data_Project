# List Comprehension

```python
numbers = [x for x in range(100)]
numbers
```

```python
# alternative (harder) way to create the above list

numbers_1 = []

for num in range(100):
  numbers_1.append(num)

numbers_1
```

```python
numbers = [x*2 for x in range(10)]
numbers
```

```python
numbers = [float(x) for x in range(10)]
numbers
```

```python
numbers = [x*2 for x in 'PYTHON']
numbers
```

```python
numbers = {x*2 for x in 'PYTHON'}   # using a set
numbers
```

```python
even_num = [x for x in range(100) if x % 2 == 0]
even_num
```
